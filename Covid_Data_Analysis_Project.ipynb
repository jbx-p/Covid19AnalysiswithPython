```python
#Remove unnecessary columns
covid_df.drop(["Sno", "Time","ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df
```

```python
# Clean the format of the date column
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%d/%m/%y')
```

```python
covid_df.head(3)
```

```python
#find the total number of active cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail
```

```python
#Let create the pivot table using the pandas library
statewise = pd.pivot_table(covid_df, values= ['Confirmed', 'Deaths', 'Cured'],
                           index = 'State/UnionTerritory', aggfunc=max)
```

```python
#Let find the recovery Rate
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
#Let find the Mortality Rate
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
#Let sort the values based on the confirmed cases column in descending order
statewise = statewise.sort_values(by = 'Confirmed', ascending= False)
```

```python
#Let now plot the pivot table
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
#let now have a look to the top 10 states based on the number of active cases

top_10_active_cases = covid_df.groupby(by= 'State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by= ['Active_Cases'], ascending= False).reset_index()
```

```python
fig = plt.figure(figsize = (16, 9))
```

```python
#Let create a bar plot
plt.title('Top 10 states with most active cases in India', size= 25)
```

```python
#Let put values into our chart
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'red')
```

```python
#let now have a look to the top 10 states based on the number of active cases

top_10_active_cases = covid_df.groupby(by= 'State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by= ['Active_Cases'], ascending= False).reset_index()
fig = plt.figure(figsize = (16, 9))
#Let create a bar plot
plt.title('Top 10 states with most active cases in India', size= 25)
#Let put values into our chart
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'red')
#Let break it down since in the above plot states names are over lapping
#let break it down

plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
#Let see the top 10 states with highest Deaths
top_10_active_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 states with most Deaths", size = 25)

ax = sns.barplot(data = top_10_active_deaths.iloc[:12], y = 'Deaths', x = "State/UnionTerritory", linewidth = 2, edgecolor = 'blue')
plt.xlabel('States')
plt.ylabel('Total Death Cases')
plt.show()
```

```python
#Let create a line chart to see the growth of active cases for top 5 states with the most number of confirmed cases
#Growth trend

fig = plt.figure(figsize=(12, 6))

ax = sns.lineplot(
    data=covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra", "Karnataka", "Kerala", "Tamil Nadu", "Uttar Pradesh"])],
    x="Date",
    y="Active_Cases",
    hue="State/UnionTerritory",
)
ax.set_title("Top 5 Affected States in India", size=16)
```

```python
#let drop a few missing values columns
vaccination = vaccine_df.drop(columns= ['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)' ], axis=1)
```

```python
#let drop all the columns where  state is india

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated': 'Total'}, inplace=True)
```

```python
#let find the State with most people vaccinated

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize= (10,5))
plt.title('Top 5 Vaccinated States in India', size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth=2, edgecolor= 'red')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

```python
#let find the State with less people vaccinated

min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total')[:5]
min_vac
```

```python
fig = plt.figure(figsize= (10,5))
plt.title('Lasr 5 Vaccinated States in India', size = 20)
x = sns.barplot(data = min_vac.iloc[:10], y = min_vac.Total, x = min_vac.index, linewidth=2, edgecolor= 'red')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
